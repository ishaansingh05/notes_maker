```python
def extract_text_from_pdf(pdf_path):
    text = ""

    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text()

    return text


# Extract text
document_text = extract_text_from_pdf(pdf_path)

print("Extraction Complete!")
print("Total Characters:", len(document_text))
```

```python
import re

def clean_text(text):
    # Remove trailing spaces from each line
    text = "\n".join(line.strip() for line in text.splitlines())

    # Replace multiple spaces/tabs with a single space
    text = re.sub(r"[ \t]+", " ", text)

    # Reduce excessive blank lines to at most two
    text = re.sub(r"\n{3,}", "\n\n", text)

    return text.strip()

cleaned_text = clean_text(document_text)

print(cleaned_text[:3000])
```

```python
import ollama
from IPython.display import display, Markdown

def summarize_chunk(chunk: str) -> str:
    system_prompt = (
        "You are a structured note-taking assistant for exam revision. "
        "You ONLY respond in strict Markdown. "
        "NEVER write plain paragraphs or prose. "
        "Always use # headings, ## subheadings, bullet points, **bold keywords**, "
        "`code/formula` backticks, ASCII diagrams, and a summary table. "
        "Violating this format means you have failed the task."
    )

    user_prompt = f"""
You are an expert professor creating HIGH-YIELD exam revision notes for university students.

Convert the given study material into structured exam notes.

══════════════════════════════════════
MANDATORY OUTPUT FORMAT — FOLLOW EXACTLY
══════════════════════════════════════

# 🎯 MAIN TOPIC TITLE HERE

## 📌 Subtopic Name

- **Term**: clear definition or key fact
- **Term**: clear definition or key fact
- Formula (if any): `formula here`

## 📌 Another Subtopic

- **Term**: key fact
- **Term**: key fact

## 🔁 Diagram (ONLY if a process, flow, or architecture is present)

## ⚡ Quick Recall Table

| Term | Meaning |
|------|---------|
| Term1 | One-line meaning |

IRON RULES:
1. START with a single # H1 heading
2. USE ## H2 for every subtopic
3. BOLD every keyword like **this**
4. BULLET POINTS only — no paragraphs ever
5. FORMULAS in backticks `like this`
6. END with the Quick Recall table always
7. MAX 12 bullet points total
8. NO prose. NO paragraphs. If you write a sentence outside a bullet, you have FAILED.

STUDY MATERIAL:
{chunk}
"""

    response = ollama.chat(
        model="qwen2.5:7b",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_prompt},
        ]
    )
    return response["message"]["content"]


def make_notes(chunk: str):
    result = summarize_chunk(chunk)
    display(Markdown(result))
```

```python
chunk_summaries = []

for i, chunk in enumerate(chunks):
    print(f" Processing chunk {i+1}/{len(chunks)}...")
    summary = summarize_chunk(chunk)       # ← summarizing happens here
    chunk_summaries.append(summary)
    display(Markdown(summary))             # ← renders it beautifully
    display(Markdown("---"))               # ← divider between chunks

print("✅ Done!")
```

```python
final_notes = "\n\n".join(chunk_summaries)
print(final_notes)
```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.pagesizes import A4

def save_pdf(text, filename="exam_notes.pdf"):
    doc = SimpleDocTemplate(filename, pagesize=A4)

    styles = getSampleStyleSheet()
    style = styles["Normal"]

    content = []

    for line in text.split("\n"):
        line = line.strip()
        if line:
            content.append(Paragraph(line, style))
            content.append(Spacer(1, 6))

    doc.build(content)
    print("PDF saved as:", filename)

save_pdf(final_notes)
```
